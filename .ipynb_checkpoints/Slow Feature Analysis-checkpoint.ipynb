### Реализация алгоритма SFA вручную

```python
import numpy as np
import pandas as pd

np.random.seed(1)
pd.set_option('display.float_format', '{:.4f}'.format)

data_size = 500
# 4 периода синуса
x = np.linspace(0, 4 * np.pi, data_size) 

sin_wave = np.sin(x)
# Создание синуса с белым шумом
white_noise = np.random.normal(0, 0.2, data_size)
series_1 = sin_wave + white_noise

# Создание синуса угасающим розовым шумом
pink_noise = (lambda x: x / np.std(x))(
    np.real(
        np.fft.ifft(
            np.concatenate([
                [0],
                (1 / np.sqrt(np.abs(np.fft.fftfreq(data_size)[1:]) + 1e-10))
                * np.exp(2j * np.pi * np.random.random(data_size - 1))
            ])
        )
    )
)

decay_envelope = 2.25 * np.exp(-2.25 * x / x.max()) + 0.25

pink_noise = pink_noise * decay_envelope

series_2 = sin_wave + 0.4 * pink_noise

# Создание синуса с выбросами и броуновским движением
brownian = np.cumsum(np.random.normal(0, 0.5, data_size))
extreme_values = np.zeros(data_size)
outlier_indices = np.random.choice(data_size, size=int(0.5 * data_size), replace=False)
extreme_values[outlier_indices] = np.random.uniform(-1, 1, size=len(outlier_indices))
series_3 = sin_wave + extreme_values + 0.4 * brownian

df = pd.DataFrame({
    'series_0': sin_wave,
    'series_1': series_1,
    'series_2': series_2,
    'series_3': series_3
})
```

```python
import matplotlib.pyplot as plt


fig, ax = plt.subplots(figsize=(12, 8))

ax.plot(x, df['series_0'], label='Синус', linewidth=1, color='#D81B60')
ax.plot(x, df['series_1'], label='Синус + белый шум', linewidth=0.8, color='#FFC107')
ax.plot(x, df['series_2'], label='Синус + розовый шум', linewidth=0.8, color='#004D40')
ax.plot(x, df['series_3'], label='Синус + выбросы + броуновское движение', linewidth=0.8, color='#A29BFE')


ax.grid(True, alpha=0.3)


ax.legend(loc='best', fontsize=11)

plt.tight_layout()

plt.show()
```

```python
# Удаляем синус, чтобы он не мешал в нашей задаче

df = df.drop('series_0', axis=1)

df.head()
```

```python
#%matplotlib notebook

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(df['series_1'], df['series_2'], df['series_3'], alpha=0.7, s=60, c='#D81B60')

ax.set_xlabel('Series 1')
ax.set_ylabel('Series 2')
ax.set_zlabel('Series 3')

plt.show()
```

```python
# Центрируем исходные данные

df_centr = pd.DataFrame()

for column in df.columns:
    df_centr[column] = df[column] - df[column].mean()
    
df_centr.head()
```

```python
# Оба способа идентичны

#A = np.cov(df_centr.T)

#A


A = (df_centr.T @ df_centr) / (len(df_centr) - 1)

A
```

```python
# Выполняем сферизацию данных

U, D, Vt = np.linalg.svd(A)
D_inv_sqrt = np.diag(D ** (-0.5))
A_s = U @ D_inv_sqrt @ U.T
df_sphered = df_centr @ A_s

df_sphered = df_sphered.rename(
    columns={0: 'series_1', 1: 'series_2', 2: 'series_3'}
)

df_sphered.head()
```

```python
#%matplotlib notebook

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt


fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(df_sphered['series_1'], df_sphered['series_2'], df_sphered['series_3'], alpha=0.7, s=60, c='#D81B60')

ax.set_xlabel('Series 1')
ax.set_ylabel('Series 2')
ax.set_zlabel('Series 3')

plt.show()
```

```python
# Вычисляем производные

df_derivative = df_sphered.diff().dropna().reset_index(drop=True)

df_derivative.head()
```

```python
# Центрируем сферированные производные

for column in df_derivative.columns:
    df_derivative[column] = df_derivative[column] - df_derivative[column].mean()
    
df_derivative.head()
```

```python
# Оба способа идентичны

#B = np.cov(df_derivative.T)

#B

B = (df_derivative.T @ df_derivative) / (len(df_derivative) - 1)

B
```

```python
lambdas, w = np.linalg.eigh(B)

print(f"Собственные числа: {lambdas}")
```

```python
sort_idx = np.argsort(lambdas)
lambdas_sorted = lambdas[sort_idx]
w_sorted = w[:, sort_idx]
# эвристика, которая нужна для воспроизводимости, так как в некоторых случаях из-за особенностей вычисления график будет отражаться
# например библиотека sksfa может вернуть перевёрнутый график по сравнению с ручной реализацией.
for i in range(w_sorted.shape[1]):
    if np.sum(w_sorted[:, i]) < 0:
        w_sorted[:, i] *= -1

w_sorted
```

```python
y = df_sphered @ w_sorted

y = y.rename(
    columns={0: 'slow_feature_1', 1: 'slow_feature_2', 2: 'slow_feature_3'}
)

y.head()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(12, 8))

ax.plot(x, sin_wave, label='Синус', linewidth=1.5, color='#D81B60')
ax.plot(x, y['slow_feature_1'], label='Первая медленная компонента', linewidth=1.5, color='#004D40')
ax.plot(x, y['slow_feature_2'], label='Вторая медленная компонента', linewidth=0.7, color='#FFA500')
ax.plot(x, y['slow_feature_3'], label='Третья медленная компонента', linewidth=0.7, color='#6495ED')
ax.grid(True, alpha=0.3)

ax.legend(loc='best', fontsize=11)

plt.tight_layout()

plt.show()
```

```python
w_original = A_s @ w_sorted

weights_df = pd.DataFrame(
    w_original, 
    index=df_sphered.columns,  
    columns=[f'Веса для медленного признака {i+1}' for i in range(w_original.shape[1])]
)

print("Вклад исходных признаков в каждую медленную компоненту:")
display(weights_df)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.bar(df_sphered.columns, np.abs(w_original[:, 0]), color='#D81B60')
plt.ylabel('Модуль весов series для первого компонентa')
plt.grid(True, alpha=0.3)
plt.show()
```

### Реализация алгоритма SFA с помощью библиотеки sksfa

```python
from sksfa import SFA
import numpy as np
import pandas as pd

sfa = SFA(n_components=3)

y_sksfa = sfa.fit_transform(df.values)


w, b = sfa.affine_parameters()

# Чтобы график компонент не был перевернутым, а веса с противоположным знаком, умножу w и y_sksfa на -1
w[0, :] *= -1
w[2, :] *= -1
y_sksfa[:,0] *= -1
y_sksfa[:,2] *= -1

weights_df = pd.DataFrame(
    w.T,
    index=df.columns,
    columns=[f'Веса для медленного признака {i+1}' for i in range(w.shape[1])]
)

print("Вклад исходных признаков в каждую медленную компоненту (sksfa):")
display(weights_df)
```

```python
print(f"Собственные числа: {sfa.delta_values_}")
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(12, 8))

ax.plot(x, sin_wave, label='Синус', linewidth=1.5, color='#D81B60')
ax.plot(x, y_sksfa[:,0], label='Первая медленная компонента (sksfa)', linewidth=1.5, color='#004D40')
ax.plot(x, y_sksfa[:,1], label='Вторая медленная компонента (sksfa)', linewidth=0.7, color='#FFA500')
ax.plot(x, y_sksfa[:,2], label='Третья медленная компонента (sksfa)', linewidth=0.7, color='#6495ED')
ax.grid(True, alpha=0.3)

ax.legend(loc='best', fontsize=11)

plt.tight_layout()

plt.show()
```
